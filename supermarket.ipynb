```python
plt.figure(figsize=(10,5))
sns.countplot(df['Ship Mode'])
plt.show()
```

#### most of the customer uses standard ship mode whereas lower number of customers took delivery .

```python
plt.figure(figsize=(10,5))
sns.countplot(df['Segment'])
plt.show()
```

#### most of the customers of superstore is consumer type i.e  purchases goods and services for personal use. There is  average number of corporate customers

```python
df['City'].value_counts().sort_values(ascending=False).head(10).plot(kind='bar')
plt.show()
```

#### New York City , Los Angeles , Philadelphia , San Francisco , Seattle , Houston , Chicago , Columbus , San Diego , Springfield this are cities with highest customers.

```python
df['State'].value_counts().sort_values(ascending=False).head(10).plot(kind='bar')
plt.show()
```

#### California , New York , Texas , Pennsylvania , Washington , Illinois , Ohio , Florida , Michigan , North Carolina this are states with highest customers

```python
plt.figure(figsize=(10,5))
sns.countplot(df['Region'])
plt.show()
```

#### Western region has high number of customers where as south region has least number of customers

```python
plt.figure(figsize=(10,5))
sns.countplot(df['Category'])
plt.show()
```

#### Most of purcheases from superstore are for office purpose.

```python
plt.figure(figsize=(20,5))
sns.countplot(df['Sub-Category'])
plt.show()
```

#### Binders and Papers are most purchased product.

```python
df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10).plot(kind='bar',figsize=(20,5))
plt.title("State's Sale",fontsize=20)
plt.xlabel('States' ,fontsize=20)
plt.ylabel('Sales',fontsize=20)
plt.show()
```

```python
df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10).plot(kind='bar',figsize=(20,5))
plt.title("City's Sale",fontsize=20)
plt.xlabel('City' ,fontsize=20)
plt.ylabel('sales',fontsize=20)
plt.show()
```

```python
df.groupby('Region')['Sales'].sum().sort_values(ascending=False).plot(kind='bar',figsize=(20,5))
plt.title("Region's Sale",fontsize=20)
plt.xlabel('Region' ,fontsize=20)
plt.ylabel('sales',fontsize=20)
plt.show()
```

```python
df.groupby('Segment')['Sales'].sum().sort_values(ascending=False).plot(kind='bar',figsize=(20,5))
plt.title("Segment's Sale",fontsize=20)
plt.xlabel('Segment' ,fontsize=20)
plt.ylabel('sales',fontsize=20)
plt.show()
```

```python
df.groupby('Category')['Sales'].sum().sort_values(ascending=False).plot(kind='bar',figsize=(20,5))
plt.title("Category's Sale",fontsize=20)
plt.xlabel('Category' ,fontsize=20)
plt.ylabel('sales',fontsize=20)
plt.show()
```

```python
df.groupby('State')['Sales','Profit'].sum().plot(kind='bar',figsize=(20,8))
plt.title("Sales and Profit",fontsize=20)
plt.xlabel('States' ,fontsize=20)
plt.ylabel('sum',fontsize=20)
plt.show()
```

```python
df.groupby('City')['Sales','Profit'].sum().sort_values(by=['Sales'],ascending=False).head(10).plot(kind='bar',figsize=(20,8))
plt.title("top 10 Cities by sales with their profit",fontsize=20)
plt.xlabel('City' ,fontsize=20)
plt.ylabel('sum',fontsize=20)
plt.show()
```

#### Philandelphia , Houston , Chicago , jacksonville this are cities which comes under top 10 cities based on  sales but they are in loss

```python
df.groupby('State')['Sales','Profit'].sum().sort_values(by=['Sales'],ascending=False).head(10).plot(kind='bar',figsize=(20,8))
plt.title("top 10 State by sales with their profit",fontsize=20)
plt.xlabel('State' ,fontsize=20)
plt.ylabel('sum',fontsize=20)
plt.show()
```

#### Texas , Pennsylvania , Florida , Illinois and Ohio  this are States which comes under top 10 states based on  sales but they are in loss

```python
df.groupby('Region')['Sales','Profit'].sum().sort_values(by=['Sales'],ascending=False).plot(kind='bar',figsize=(20,8))
plt.title(" Region by sales with their profit",fontsize=20)
plt.xlabel('Region' ,fontsize=20)
plt.ylabel('sum',fontsize=20)
plt.show()
```

#### West region has highest sales and profit but south region has good sales and profit ratio.

```python
df.groupby('Category')['Sales','Profit'].sum().sort_values(by=['Sales'],ascending=False).plot(kind='bar',figsize=(20,8))
plt.title(" Category by sales with their profit",fontsize=20)
plt.xlabel('Category' ,fontsize=20)
plt.ylabel('sum',fontsize=20)
plt.show()
```

#### products from Technology category have high sales and profit.  Office supplies has good sales and profit ratio but for furniture sales-profit ratio is very low.

```python
df.groupby('City')['Sales','Profit'].sum().sort_values(by=['Profit'],ascending=False).head(10).plot(kind='bar',figsize=(20,8))
plt.title("top 10 Cities by profit with their sales",fontsize=20)
plt.xlabel('City' ,fontsize=20)
plt.ylabel('sum',fontsize=20)
plt.show()
```

#### Atlanta and Minneapolis cities have good sales-profit ratio as compare to highly profited cities.

```python
df.groupby('State')['Sales','Profit'].sum().sort_values(by=['Profit'],ascending=False).head(10).plot(kind='bar',figsize=(20,8))
plt.title("top 10 State by profit with their sales",fontsize=20)
plt.xlabel('State' ,fontsize=20)
plt.ylabel('sum',fontsize=20)
plt.show()
```

#### Minnesota and Delaware states have good sales-profit ratio.
